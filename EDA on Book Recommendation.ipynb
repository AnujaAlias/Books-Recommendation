```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore') 
```

Importing Books dataset and convert it into data frame and display it.
  

 -DATA SOURCE : Kaggle

```python
# Import dataset
data=pd.read_csv("books.csv")
data.head(10)
```

This dataset contains following columns;
- bookID              :	Unique identifier for each book
- title               :	Title of the book
- authors             :   Author(s) of the book
- average_rating      :   Average rating of the book
- ratings_count       :	Total number of ratings for the book,Number of unique rating recieved
- text_reviews_count  :	Total number of text reviews for the book
- publication_date    :	Date of publication of the book
- publisher           :	Publisher of the book
- isbn                :   The International Standard Book Number (ISBN) is a numeric commercial book identifier that is intended
- isbn13              :  13-digit ISBNs were introduced and used to increase the availability of ISBNs worldwide written language of each book

```python
# drop the extra unamed column
data=data.drop("Unnamed: 12",axis=1)
data
```

```python
# Create dataframe and display it
df =pd.DataFrame(data)
df
```

```python
# Rename columns 

df.rename(columns = 
                {
            "bookID" : "BookID", 
            "title" : "Title", 
            "authors" : "Authors", 
            "average_rating" : "Ratings", 
            "isbn" : "ISBN",
            "isbn13" : "ISBN13",
            "language_code" : "Language",
            "ratings_count" : "Ratings_Count",
            "text_reviews_count" : "Reviews_Count",
            "publication_date" : "Date",
            "publisher" : "Publisher",
            "  num_pages" : "Pages"

                }, 
            inplace = True
               )
df.head(10)
```

```python
# convert mixed datetime formats of Date column to a single standard one 
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Date'] = df['Date'].dt.strftime('%m/%d/%Y')
df
```

```python
# Categorizing books by length

df1=df.copy()
df1
def add_book_length_category(df1):
    bins = [0, 250, 500, float('inf')]
    labels = ['Short', 'Medium', 'Long']
    df1['book_length_category'] = pd.cut(df['Pages'], bins=bins, labels=labels, right=False)
add_book_length_category(df1)




# Group the data by the 'book_length_category' column and count the occurrences
category_counts = df1['book_length_category'].value_counts()

# Create a pie chart
plt.figure(figsize=(7, 7))
mylabels = ["Short", "Medium", "Long"]
myexplode = (0.09, 0, 0)
plt.pie(category_counts, labels=mylabels,autopct='%1.1f%%',explode=myexplode ,startangle=140,shadow = True)
plt.title("Book Length Category Distribution")
plt.legend(title="Categories")
plt.show()
```

```python
 
max_reviewscount_title = df.groupby('Title')['Reviews_Count'].mean()

top_10_books_reviews = max_reviewscount_title.sort_values(ascending=False).head(10)
top_10_books_reviews
df4=pd.DataFrame(top_10_books_reviews)

plt.figure(figsize=(10,5))
sns.lineplot(x='Title',y='Reviews_Count',data=df4)
sns.set_context("paper")
sns.set_palette('vlag')
plt.title("Lineplot of Books Based on Review Count")
plt.xlabel("Title")
plt.ylabel("Reviews_count")


plt.xticks(rotation=90)
plt.show
```

```python

# scatter plot



# Group the data by 'Authors' and calculate the maximum rating for each author
max_ratingscount_author = df.groupby('Authors')['Ratings_Count'].max()

# Sort the authors by their maximum ratings in descending order
top_20_authors = max_ratingscount_author.sort_values(ascending=False).head(20)
top_20_authors


# Group the data by 'Titls' and calculate the maximum rating count for each author
max_ratingscount_Title = df.groupby('Title')['Ratings_Count'].max()

# Sort the authors by their maximum ratings in descending order
top_20_Titles = max_ratingscount_Title.sort_values(ascending=False).head(20)
top_20_Titles

# plot 1
plt.subplots(figsize=(30,4))
plt.subplot(1, 2, 1)
plt.scatter(top_20_authors.index,top_20_authors.values,linewidth=2.5,c='hotpink')
plt.title('Scatter Plot of Top 20 Authors based on Rating Count')
plt.xlabel('Authors')
plt.ylabel('Ratings_Count')
plt.xticks(rotation=90)


# plot 2

plt.subplots(figsize=(30,4))
plt.subplot(1, 2, 2)
plt.scatter(top_20_Titles.index,top_20_Titles.values,linewidth=2.5,c='blue')
plt.title('Scatter Plot of Top 20 Book Titles based on Rating Count')
plt.xlabel('Title')
plt.ylabel('Ratings_Count')
plt.xticks(rotation=90)
plt.show()

```
